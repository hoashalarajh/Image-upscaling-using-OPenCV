# Image Upscaling Using OpenCV

### Using the INTER_CUBIC Interpolation Function in OpenCV

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# reading the image and displaying the image
# reading the image
img_input = cv.imread("thalapathy_beast.png")
# converting the color from BGR to RGB
img_input = cv.cvtColor(img_input, cv.COLOR_BGR2RGB)
print (img_input.shape)
plt.imshow(img_input)
plt.axis("off")
plt.show()
```

```python
# image resizing
resized_img = cv.resize(img_input, (1920, 1080), interpolation = cv.INTER_CUBIC)
# getting the size of the resized image
print (resized_img.shape)
```

```python
# plotting the resized image
plt.imshow(resized_img)
plt.axis("off")
plt.show()
```
